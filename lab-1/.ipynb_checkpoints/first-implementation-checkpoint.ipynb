```python
import os
import time
from collections import Counter
from pprint import pprint
```

```python
# folder_name = input("Fill in the folder name in current directory")
path = os.path.abspath('documents')
os.chdir(path)
```

```python
def count_unique_words(str, counter):
    counted_words = Counter(str.split('\n'))
    counter += counted_words

def read_dir_contents():
    unique_words_counter = Counter()
    
    def read_file(file_path):
        with open(file_path, 'r') as f:
            return f.read()
         
    for file in os.listdir():
        count_unique_words(read_file(f"{path}\{file}"), unique_words_counter)
    return unique_words_counter
```

```python
start_time = time.time()
pprint(read_dir_contents())
print("Time elapsed: %s sec" % (time.time() - start_time))
```
